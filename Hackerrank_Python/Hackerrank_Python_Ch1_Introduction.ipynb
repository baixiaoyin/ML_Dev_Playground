## Q6: Function

### Question

```python
image_path = 'hackerrank_python_img/Intro5_Function.jpg'
display(Image(filename=image_path, width=400))
```

### Solution

```python
def is_leap(year):
    leap = False
    
    # Write your logic here
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        leap = True    
    return leap

year = int(input())
print(is_leap(year))
```
