### Solution

```python
if __name__ == '__main__':

    def handler(result):
        inp = input().split()
        command = inp[0]
        values = inp[1:]
        if command == 'print':
            print(result)
        else:
            execute = 'result.' + command + "(" + ",".join(values) + ")"
            eval(execute)
    
    
    result = []
    for i in range(int(input())):
        handler(result)
```

## Q3: List Comprehension

### Question

```python
image_path = 'hackerrank_python_img/BasicDataType2_ListComprehensions.jpg'
display(Image(filename=image_path, width=400))
```

### Solution

```python
if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
    
    output = [[i, j, k] for i in range(x + 1) for j in range(y + 1) for k in range(z + 1) if i + j + k != n]
    
    print(output)
```

### Take away:
List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list. <br>
The syntax is newlist = [expression for item in iterable if condition == True]

### Question

```python
image_path = 'hackerrank_python_img/BasicDataType3_FindTheRunnerUpScore.jpg'
display(Image(filename=image_path, width=400))
```

### Solution

```python
if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    list1 = [x for x in arr if x != max(arr)]
    output = max(list1)
    print(output)
```

## Q4 Nested List

### Question

```python
image_path = 'hackerrank_python_img/BasicDataType4_NestedList.jpg'
display(Image(filename=image_path, width=800))
```

### Solution

```python
if __name__ == '__main__':
    name_score_list = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        name_score_list.append([name, score])
    
    secondlowestscore = sorted(set(s[1] for s in name_score_list))[1]
    name = sorted([n[0] for n in name_score_list if n[1]==secondlowestscore])
    for i in name:
        print(i)
```

### Take away:
1.set is a collection in Python that has no duplicates

## Q5: Find the percentage

```python
image_path = 'hackerrank_python_img/BasicDataType5_FindThePercentage.jpg'
display(Image(filename=image_path, width=400))
```

## Solution

```python
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    result = sum(student_marks[query_name])/len(student_marks[query_name])
    print('%.2f' % result)
```

### Take away:
1. name, *line = input().split(): that use *line to create a list of all elements after the first element <br>
2. How to access element in Dict: dict[key_value] <br>
2. '%.2f' % result: two decimal places of the varable result
