## Q1: Swap case

### Question

```python
image_path = 'hackerrank_python_img/String0_swapcase.jpg'
display(Image(filename=image_path, width=400))
```

### Solution

```python
def swap_case(s):
    return s.swapcase()

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

### Take away:
Python 3 string got built in swapcase() method

## Q2: String split and join

### Question

```python
image_path = 'hackerrank_python_img/String1_stringsplitandjoin.jpg'
display(Image(filename=image_path, width=400))
```

### Solution

```python
def split_and_join(line):
    # write your code here
    line = line.split()
    return "-".join(line)

if __name__ == '__main__':
    line = input()
    result = split_and_join(line)
    print(result)
```

### Take away
1. string.split(separator, maxsplit), maxsplit means that the length of splited array would be maxsplt + 1
2. string.join(iterable)
